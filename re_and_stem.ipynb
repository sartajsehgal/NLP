```python
wsj = sorted(set(nltk.corpus.treebank.words()))
```

```python
# find all decimal numbers in the wsj list
[w for w in wsj if re.search('^[0-9]+\.[0-9]+$',w)]
```

```python
# find words ending with $ sign
[w for w in wsj if re.search('^[A-Z]+\$$',w)]
```

```python
# words with exactly 4 digits
[w for w in wsj if re.search('^[0-9]{4}$',w)]
```

```python
# words starting with one or number followed by hyphen and 3 to 5 characters
[w for w in wsj if re.search('^[0-9]+-[a-z]{3,5}$',w)]
```

```python
# words starting with atleast 5 charachters followed by hyphen followed by 2-3 letters, hyphen and maximum 6 chars
[w for w in wsj if re.search('^[a-z]{5,}-[a-z]{2,3}-[a-z]{,6}$',w)]
```

```python
# words ending with ed or ing
[w for w in wsj if re.search('(ed|ing)$',w)]
```

```python
word="building"
porterstemmer = nltk.stem.PorterStemmer()
print(porterstemmer.stem(word))
```

```python
lancasterstemmer=nltk.stem.LancasterStemmer()
print(lancasterstemmer.stem(word))
```

```python
# Wordnet Lemmatizer
lemmatizer=nltk.stem.WordNetLemmatizer()
pos='n' # if it is a noun
print("Lemmatization of",word,"(",pos,"):",lemmatizer.lemmatize(word,pos))
pos='v' # if it is a verb
print("Lemmatization of",word,"(",pos,"):",lemmatizer.lemmatize(word,pos))
```

```python
# 1. Wordlist Lexical resource - It is a lexicon (collection of words) that contains set of commonly occuring words
# Words- does not include any morphological variants of words. i.e. it is a word listof base/root word forms
words.fileids()
```

```python
words.words()
```

```python
# FIltering Text - to remove morphological variants or misspelled words in text
def unusual_words(text):
    text_vocab=set([w for w in text if w.isalpha()])
    english_vocab=set([w for w in words.words()])
    unusual_words=text_vocab.difference(english_vocab)
    return unusual_words
```

```python
unusual_words(gutenberg.words('austen-emma.txt'))
```

```python
freq_letters=nltk.FreqDist('egivrvonl')
wordlist=words.words()
obligatory='r'
[w for w in wordlist if len(w)>=6 and obligatory in w and nltk.FreqDist(w) <= freq_letters]
```

```python
# 2. Stopwords - it is a list of commonly/high frequency words of language which do not have any semantic importance
stopwords.fileids()
```

```python
stopwords.words('english')
```

```python
def fraction_nostopword(text):
    stopword=stopwords.words(fileids='english')
    content_non_stopword=[w for w in text if w not in stopword]
    return len(content_non_stopword)/len(text)
```

```python
fraction_nostopword(gutenberg.words(fileids='austen-emma.txt'))
```

```python
cfd=nltk.ConditionalFreqDist(
    (fileid,name[-1])
    for fileid in names.fileids()
    for name in names.words(fileid)
)
```

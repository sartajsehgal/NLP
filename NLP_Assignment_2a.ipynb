```python
# 1. In the news genre of brown corpus, find the count of the words starting with wh, such as what, when, where, who and why?
from nltk.probability import FreqDist
vec=[w for w in brown.words(categories='news') if w.startswith('wh')]
fdist=FreqDist(vec)
print(len(vec))
fdist
```

```python
# 2. Find the conditional frequency distribution of modals ['can', 'could', 'may', 'might', 'must', 'will'] in all the categories of brown corpus?
from nltk.probability import ConditionalFreqDist
modals=['can', 'could', 'may', 'might', 'must', 'will']
category=brown.categories()
cfd = ConditionalFreqDist(
    (cat,w)
    for cat in category
    for w in brown.words(categories=cat)
    if w in modals)
cfd.tabulate(condition=category, samples=modals)
```

```python
# 4. Read in the texts of the State of the Union addresses, using the state_union corpus reader. Count occurrences of men , women , and people in each document. What has happened to the usage of these words over time? 
documents=state_union.fileids()
mods=['men','women','people']
cfd1 = ConditionalFreqDist(
    (doc,w)
    for doc in documents
    for w in state_union.words(doc)
    if w.lower() in mods)
cfd1.tabulate(condition=documents, samples=mods)
# Usage of these words firstly decreased and then increased over time
```

```python
# 7. Write a program to generate a table of lexical diversity scores (i.e., token/type ratios) for each genre. Include the full set of Brown Corpus genres ( nltk.corpus.brown.categories() ). Which genre has the lowest diversity 
lex_div_scores={}
category=brown.categories()
for cat in category:
    words=brown.words(categories=cat)
    lex_div_scores[cat]=len(words)/len(set(words))
min_val=min(lex_div_scores.values())
min_val_key=[key for key in lex_div_scores.keys() if lex_div_scores[key]==min_val]
print('The category with min value:', min_val_key)
lex_div_scores
```

```python
def most_freq_50(text):
    fdist_text=FreqDist(text.words())
    return fdist_text.most_common(50)
```

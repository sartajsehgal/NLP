```python
# Ngrams- Collection of n words, e.g for n=2 bigram, n=3 trigram etc
# It is done because many models take ngrams as input
sentence = "I am Sartajbir Singh, I am a third year computer engineering student"
from nltk import ngrams
n=2
n_grams=ngrams(sentence.split(),n)
```

```python
for gram in n_grams:
    print(gram)
```

```python
# Collocations - The words that occur together quite often
from nltk.collocations import *
bigram_measures = nltk.collocations.BigramAssocMeasures()
finder = BigramCollocationFinder.from_words(text1) # made bigrams from text1
finder.nbest(bigram_measures.pmi,100) # found 100 best collocations using pmi as metric
```

```python
# stylistics - to study systematic differences between different genres
modals=['can','could','would','will','may','might','must']
category=brown.categories()
for cat in category:
    words=brown.words(categories=cat)
    fdist=FreqDist([w.lower() for w in words])
    for modal in modals:
        print(modal,fdist[modal],cat)
```
